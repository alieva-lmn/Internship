# 2. Check if there are any missing values.

```python
missing_values = df.isna() 

# isna() returns a DataFrame or Series of 
#the same shape as the original but with Boolean values indicating
# whether each value is missing (True) or not (False).

if missing_values.any().any():
    print("There are missing values.")
else:
    print("There are no missing values.")

# any().any(): Checks if there are any missing values in the entire DataFrame

missing_counts = missing_values.sum()
print("\nMissing value counts per column:")
print(missing_counts)
```
